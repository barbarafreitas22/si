1.4) Select all samples from the dataset with values less than or equal to 6 for all independent variables/features. How many samples do you obtain?

```python
# Creating a Boolean mask to fit the condition (all features <= 6)
mask = iris_dataset.X <= 6

# Identify the samples where all feautures satisfy the condition
selected_samples = np.all(mask, axis=1)

# Select the respective labels (samples that meet the condition
samples_meeting_condition = iris_dataset.X[selected_samples]
total_samples  = samples_meeting_condition.shape[0]

print(f"Total of samples where all independent variables are ≤ 6: {total_samples}")
```
