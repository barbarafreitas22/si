## Exercise 2: NumPy array Indexing/Slicing

Examples of how to use the methods in exercise 2

```python
import numpy as np
from si.data.dataset import Dataset
```

```python
# --- Setup with examples ---
X = np.array([[1, 2, np.nan],
              [3, 4, 5],
              [np.nan, 6, 7]])
y = np.array([10, 20, 30])

print("\n--- Example fillna ('mean') ---")
dataset = Dataset(X=np.copy(X), y=np.copy(y), label="Y")

print("Features values with missing values:\n", dataset.X)
print("Labels with missing values:", dataset.y)
print("Shape of X with missing values:", dataset.X.shape)
print("Shape of Y with missing values:", dataset.y.shape)

# Apply fillna with "mean"
dataset = dataset.fillna("mean")

print("Shape of X after fillna:", dataset.X.shape)
print("Shape of y after fillna:", dataset.y.shape)
print("Features values without missing values (filled):\n", dataset.X)
print("Labels after fillna:", dataset.y)



print("\n--- Exemplo fillna (valor 0) ---")
dataset = Dataset(X=np.copy(X), y=np.copy(y), label="Y")

print("Features values with missing values:\n", dataset.X)


# # Apply fillna with 0
dataset = dataset.fillna(0)

print("Features values without missing values (filled with 0):\n", dataset.X)


# ---------------------------------------------------
# Example remove_by_index
# ---------------------------------------------------
print("\n--- Exemplo remove_by_index (index 1) ---")
dataset = Dataset(X=np.copy(X), y=np.copy(y), label="Y")

print("Features values before removal:\n", dataset.X)
print("Labels before removal:", dataset.y)
print("Shape of X before removal:", dataset.X.shape)
print("Shape of Y before removal:", dataset.y.shape)

# Remove sample at index 1 
dataset = dataset.remove_by_index(1)

print("Shape of X after removal:", dataset.X.shape)
print("Shape of y after removal:", dataset.y.shape)
print("Features values after removal:\n", dataset.X)
print("Labels after removal:", dataset.y)
```
