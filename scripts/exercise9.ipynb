```python
import numpy as np
# Required custom imports
from si.io.csv_file import read_csv 
from si.model_selection.split import stratified_train_test_split 
from si.models.random_forest_classifier import RandomForestClassifier
from si.metrics.accuracy import accuracy 

# --- 1. Use the iris dataset ---
path_to_iris = '../datasets/iris/iris.csv' 
iris_dataset = read_csv(path_to_iris, sep=',', label=True)

print("--- Random Forest Classifier Test Protocol ---")
print("Iris Dataset loaded. Shape:", iris_dataset.shape())
print("-" * 30)

# --- 2. Split the data into train and test sets ---
TEST_SIZE = 0.3
RANDOM_STATE = 42

# Use stratified split to ensure proportion of classes is maintained
train_set, test_set = stratified_train_test_split(
    iris_dataset, 
    test_size=TEST_SIZE, 
    random_state=RANDOM_STATE
)

print(f"Split completed. Train: {train_set.shape()[0]} samples | Test: {test_set.shape()[0]} samples")
print("-" * 30)

# --- 3. Create the RandomForestClassifier model ---
N_ESTIMATORS = 100 # Number of trees
MAX_DEPTH = 5      # Maximum tree depth
MAX_FEATURES = 2   # Standard heuristic for feature selection (sqrt(n_features) for Iris)

rf_model = RandomForestClassifier(
    n_estimators=N_ESTIMATORS,
    max_depth=MAX_DEPTH,
    max_features=MAX_FEATURES
)

# --- 4. Train the model ---
rf_model.fit(train_set)

print(f"Model Trained: Random Forest with {N_ESTIMATORS} trees.")
print("-" * 30)


# --- 5. What is the score of the model on the test set? ---
test_accuracy = rf_model.score(test_set)

print(f"Final Result:")
print(f"Accuracy on Test Set: {test_accuracy:.4f}")
```
